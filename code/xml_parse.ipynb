```python
all_english_texts = []
all_anchor_list = []
all_lemma_list = []
all_BabelnetIDs = [] 

# defined Python user-defined exceptions
class Error(Exception):
    """Base class for other exceptions"""
    pass

class EmptyTagError(Error):
    """Raised when the text lang tag is empty"""
    print("Empty tag detected")
    next

# parsing code
file_path = '/Users/oluwayetty1/Desktop/School/1-2/NLP/public_homework2/test.xml'
xml_content = ET.iterparse(file_path, events=('end',), tag='sentence')
for event, element in xml_content:
    anchor = []
    lemma = []
    babelnetID = []
    try:
        for elem in element.iter():
            if elem.text == None and elem.attrib['lang'] == 'en' and elem.text == None:
                raise EmptyTagError()
            else:
                if elem.tag == 'text' and elem.attrib['lang'] == 'en' and elem.text != None:
                    all_english_texts.append(elem.text)
                if elem.tag == 'annotations':
                    for annotation in elem.iter():
                        if annotation.tag == 'annotation' and annotation.attrib['lang']  == 'en':
                            lemma.append(annotation.attrib['lemma'])
                            anchor.append(annotation.attrib['anchor'])
                            babelnetID.append(annotation.text.strip())
        all_BabelnetIDs.append(babelnetID)
        all_lemma_list.append(lemma)
        all_anchor_list.append(anchor)
    except EmptyTagError:
        next 
    element.clear()
    while element.getprevious() is not None:
        del element.getparent()[0]
```

```python
def write_per_row(filename, list_, header_name):
    with open(filename, mode='w') as csv_file:
        fieldnames = [header_name]
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()
        for element in list_:
            writer.writerow({header_name: element})

write_per_row('/Users/oluwayetty1/Desktop/School/1-2/NLP/public_homework2/data/sentence.csv',all_english_texts, 'sentence')
write_per_row('/Users/oluwayetty1/Desktop/School/1-2/NLP/public_homework2/data/anchor.csv', all_anchor_list, 'anchor')
# write_per_row('/Users/oluwayetty1/Desktop/School/1-2/NLP/public_homework2/data/lemma.csv',all_lemma_list,'lemma')
write_per_row('/Users/oluwayetty1/Desktop/School/1-2/NLP/public_homework2/data/lemma_id.csv',arr, 'lemma_id')
```

```python
def tokenize(features):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(features)
    corpus = tokenizer.texts_to_sequences(features)
    nb_samples = sum(len(s) for s in corpus)
    V = len(tokenizer.word_index) + 1
    return corpus, V
```

```python
def generate_data(corpus, V):
    for words in corpus:
        contexts = []
        labels   = []            
        contexts.append(words[0])
        labels.append(words[1])

        x = sequence.pad_sequences(contexts, maxlen=maxlen)
        y = np_utils.to_categorical(labels, V)
        print(x, y)
        yield (x, y)

generate_data(data_corpus,feature_V)
```

```python
dim = 3

cbow = Sequential()
cbow.add(Embedding(input_dim=feature_V, output_dim=dim, input_length=maxlen))
cbow.add(Lambda(lambda x: K.mean(x, axis=1), output_shape=(dim,)))
cbow.add(Dense(feature_V, activation='softmax'))
```

```python
cbow.compile(loss='categorical_crossentropy', optimizer='adadelta')
```

```python
for ite in range(10):
    loss = 0.
    for x, y in generate_data(data_corpus, feature_V):
        loss += cbow.train_on_batch(x, y)

    print(ite, loss)
```
